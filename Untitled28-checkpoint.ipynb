```python
mygraph = {
    'A':['D','B'],
    'B':['C','G','E'],
    'C':['A'],
    'D':['C'],
    'E':['H'],
    'F':[],
    'G':['F'],
    'H':['G','F']
}
visited = []
queue = []
starting = input("Enter your Starting point: ")
ending = input("Enter your Ending point: ")
def mybfs(mygraph,visited,starting_node,ending_node):
    visited.append(starting_node)
    queue.append(starting_node)
    
    while queue:
        
        a = queue.pop(0)
        if a == ending:
            print(a,end = " ")
            break
        else:
            print(a,end = " ")
            
        for neighbour in mygraph[a]:
            if neighbour not in visited:
                visited.append(neighbour)
                queue.append(neighbour)

print("Breadth first search: ")
mybfs(mygraph,visited,starting,ending)
```

```python
graph = {
    'A':['D','B'],
    'B':['C','G','E'],
    'C':['A'],
    'D':['C'],
    'E':['H'],
    'F':[],
    'G':['F'],
    'H':['G','F']
}
visited = []
starting = input("Enter your Starting point: ")
ending = input("Enter your Ending point: ")
def bfs_with_cost(graph, start, goal):
    queue = deque([(start, [start], 0)])
    visited = set()
    order_of_expansion = []
    
    while queue:
        node, path, cost = queue.popleft()
        if node == goal:
            return order_of_expansion, cost
        
        visited.add(node)
        order_of_expansion.append(node)

        for neighbor, neighbor_cost in graph[node]:
            if neighbor not in visited:
                queue.append((neighbor, path + [neighbor], cost + neighbor_cost))

    return None, None
```
